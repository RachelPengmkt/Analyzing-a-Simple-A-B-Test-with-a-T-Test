```python
# Load data
df_e = pd.read_csv('Experiment Data.csv', skiprows=1, usecols = ['Impressions','Clicks','Conversions'])
df_c = pd.read_csv('Control Data.csv', skiprows=1, usecols = ['Impressions','Clicks','Conversions'])
```

```python
# drop last 5 rows
df_e.drop(df_e.tail(5).index,inplace=True) 
df_c.drop(df_c.tail(5).index,inplace=True) 
```

```python
#Calcualte CTR and Conversion Rate for both groups and append the data to original data frame
df_e['CTR'] = df_e['Clicks'].astype(float)/df_e['Impressions'].astype(float)
df_c['CTR'] = df_c['Clicks'].astype(float)/df_c['Impressions'].astype(float)
df_e['Conversion Rate'] = df_e['Conversions'].astype(float)/df_e['Clicks'].astype(float)
df_c['Conversion Rate'] = df_c['Conversions'].astype(float)/df_c['Clicks'].astype(float)
```

```python
#Check out updated data of Experiment group and sort by conversions
df_e.sort_values('Conversions', ascending=False)
```

```python
#Check out updated data of Control group and sort by conversions
df_c.sort_values('Conversions', ascending=False)
```

```python
#Replace NaN with 0
df_e['Conversion Rate'] = df_e['Conversion Rate'].fillna(0)
df_c['Conversion Rate'] = df_c['Conversion Rate'].fillna(0)
df_e['CTR'] = df_e['CTR'].fillna(0)
df_c['CTR'] = df_c['CTR'].fillna(0)
```

```python
#Check out updated data of Control group
df_c.sort_values('Conversions', ascending=False)
```

```python
#Check out updated data of Control group
df_e.sort_values('Conversions', ascending=False)
```

```python
#Calculate Mean Conversion Rate and Mean CTR for Experiment group
SumConv_e = df_e['Conversions'].astype(float).sum()
SumClick_e = df_e['Clicks'].astype(float).sum()
SumImpr_e = df_e['Impressions'].astype(float).sum()
MeanCR_e = SumConv_e / SumClick_e
MeanCTR_e = SumClick_e / SumImpr_e
```

```python
#Calculate Mean Conversion Rate and Mean CTR for Control group
SumConv_c = df_c['Conversions'].astype(float).sum()
SumClick_c = df_c['Clicks'].astype(float).sum()
SumImpr_c = df_c['Impressions'].astype(float).sum()
MeanCR_c = SumConv_c / SumClick_c
MeanCTR_c = SumClick_c / SumImpr_c
```

```python
#Print out Mean Conversion Rate and Mean CTR for both group
print('Mean Conversion Rate Experiment', MeanCR_e)
print('Mean Conversion Rate Control', MeanCR_c)
print('Mean CTR Experiment', MeanCTR_e)
print('Mean CTR Control', MeanCTR_c)
```

```python
#Calculate Standard Deviation of Conversion Rate and CTR for both group
SdCR_e = np.sqrt(MeanCR_e*(1- MeanCR_e) )
SdCR_c = np.sqrt(MeanCR_c*(1 - MeanCR_c))
SdCTR_e = np.sqrt(MeanCTR_e*(1-MeanCTR_e))
SdCTR_c = np.sqrt(MeanCTR_c*(1-MeanCTR_c))
```

```python
#Print out Standard Deviation of Conversion Rate and CTR for both group
print('SD Conversion Rate Experiment', SdCR_e)
print('SD Conversion Rate Control', SdCR_c)
print('SD CTR Experiment', SdCTR_e)
print('SD CTR Control', SdCTR_c)
```

```python
# Calculate the t-statistics and p value of Conversion Rate
# Assuming unequal variance
s_CR = np.sqrt( SdCR_e*SdCR_e / SumClick_e + SdCR_c*SdCR_c / SumClick_c )
t_CR =  (MeanCR_e-MeanCR_c) / (s_CR  ) 
df_CR = s_CR **2 /( (SdCR_e*SdCR_e / SumClick_e)**2 / (SumClick_e-1) + (SdCR_c*SdCR_c / SumClick_c)**2 / (SumClick_c-1) )
print('t statistics of t-test on Conversion Rate', t_CR)
print('Degree of freedom of t-test on Conversion Rate', df_CR)

from scipy import stats
p_value = 2*(1-stats.t.cdf(abs (t_CR),df_CR-1))
print('p value of t-test on Conversion Rate', p_value)
```

```python
# Calculate the t-statistics and p value of CTR
# Assuming unequal variance
s_CTR = np.sqrt( SdCTR_e*SdCTR_e / SumImpr_e + SdCTR_c*SdCTR_c / SumImpr_c )
t_CTR =  (MeanCTR_e-MeanCTR_c) / (s_CTR  ) 
df_CTR = s_CTR **2 /( (SdCTR_e*SdCTR_e / SumImpr_e)**2 / (SumImpr_e-1) + (SdCTR_c*SdCTR_c / SumImpr_c)**2 / (SumImpr_c-1) )
print('t statistics of t-test on CTR', t_CTR)
print('Degree of freedom of t-test on CTR', df_CTR)
from scipy import stats
p_value = 2*(1-stats.t.cdf(abs (t_CTR),df_CTR-1))
print('p value of t-test on CTR', p_value)
```
