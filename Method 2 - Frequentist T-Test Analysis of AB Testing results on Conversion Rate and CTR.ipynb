```python
# Import the packages
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# Load data
df_e = pd.read_csv('Experiment Data.csv', skiprows=1, usecols = ['Impressions','Clicks','Conversions'])
df_c = pd.read_csv('Control Data.csv', skiprows=1, usecols = ['Impressions','Clicks','Conversions'])
```

```python
# Drop last 5 rows
df_e.drop(df_e.tail(5).index,inplace=True) 
df_c.drop(df_c.tail(5).index,inplace=True) 
```

```python
# Calculate the total amount of Conversions, clicks and impressions for Experiment group
SumConv_e = int (df_e['Conversions'].astype(float).sum())
SumClick_e = int (df_e['Clicks'].astype(float).sum())
SumImpr_e = int( df_e['Impressions'].astype(float).sum())
```

```python
# Print out the total amount of Conversions, clicks and impressions for Experiment group
print('Total Conversions of Experiment Group', SumConv_e)
print('Total Clicks of Experiment Group', SumClick_e)
print('Total Impressions of Experiment Group', SumImpr_e)
```

```python
# Calculate the total amount of Conversions, clicks and impressions for Control group
SumConv_c = int (df_c['Conversions'].astype(float).sum())
SumClick_c = int (df_c['Clicks'].astype(float).sum())
SumImpr_c = int( df_c['Impressions'].astype(float).sum())
```

```python
# Convert data in Experiment group to 0, 1
dfE_1 = pd.DataFrame(columns=['Clicks','Conversions'])
dfE_1['Clicks'] = np.ones(int (SumClick_e)).tolist() 
dfE_1['Conversions'] = np.ones(int(SumConv_e)).tolist() + np.zeros(int (SumClick_e-SumConv_e) ).tolist()
dfE_1['Conversion Rate'] = dfE_1['Conversions']/dfE_1['Clicks']
dfE_1
```

```python
# Convert data in Experiment group to 0, 1
dfE_2 = pd.DataFrame(columns=['Clicks','Impressions'])
dfE_2['Impressions'] = np.ones(int (SumImpr_e)).tolist() 
dfE_2['Clicks'] = np.ones(int(SumClick_e)).tolist() + np.zeros(int (SumImpr_e-SumClick_e) ).tolist()
dfE_2['CTR'] = dfE_2['Clicks']/dfE_2['Impressions']
dfE_2
```

```python
# Convert data in Control group to 0, 1
dfC_1 = pd.DataFrame(columns=['Clicks','Conversions'])
dfC_1['Clicks'] = np.ones(int (SumClick_c)).tolist() 
dfC_1['Conversions'] = np.ones(int(SumConv_c)).tolist() + np.zeros(int (SumClick_c-SumConv_c) ).tolist()
dfC_1['Conversion Rate'] = dfC_1['Conversions']/dfC_1['Clicks']
dfC_1
```

```python
# Convert data in Control group to 0, 1
dfC_2 = pd.DataFrame(columns=['Clicks','Impressions'])
dfC_2['Impressions'] = np.ones(int (SumImpr_c)).tolist() 
dfC_2['Clicks'] = np.ones(int(SumClick_c)).tolist() + np.zeros(int (SumImpr_c-SumClick_c) ).tolist()
dfC_2['CTR'] = dfC_2['Clicks']/dfC_2['Impressions']
dfC_2
```

```python
from scipy.stats import ttest_ind, ttest_ind_from_stats
from scipy.special import stdtr
# Use scipy.stats.ttest_ind.
t, p = ttest_ind(dfC_1['Conversion Rate'], dfE_1['Conversion Rate'], equal_var=False)
print("ttest_ind: t = %g  p = %g" % (t, p))
```

```python
# Use scipy.stats.ttest_ind.
t, p = ttest_ind(dfC_2['CTR'], dfE_2['CTR'], equal_var=False)
print("t-statistics and p-value: t = %g  p = %g" % (t, p))
```

```python
# Calculate Means of CTR in both groups
dfC_CR_Mean = dfC_1['Conversion Rate'].mean()
dfE_CR_Mean = dfE_1['Conversion Rate'].mean()
```

```python
# Print out Means of CTR in both groups
print('Mean of Conversion Rate in Control Group', dfC_CR_Mean)
print('Mean of Conversion Rate in Experiment Group', dfE_CR_Mean)
```

```python
# Calculate Means of CTR in both groups
dfC_CTR_Mean = dfC_2['CTR'].mean()
dfE_CTR_Mean = dfE_2['CTR'].mean()
```

```python
# Print out Means of CTR in both groups
print('Mean of CTR in Control Group', dfC_CTR_Mean)
print('Mean of CTR in Experiment Group', dfE_CTR_Mean)
```
